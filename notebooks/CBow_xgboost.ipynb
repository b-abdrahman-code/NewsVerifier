```python
import pandas as pd
import numpy as np
```

```python

df_full = pd.read_csv('cleaned.csv')


df_clean = df_full[['news_content_clean', 'label']].dropna()

def create_word_sentence_cbow(data):
    return [text.split() for text in data]

splited_data = create_word_sentence_cbow(df_clean['news_content_clean'])


from gensim.models import Word2Vec
Cbow = Word2Vec(
    sentences=splited_data,
    sg=0,
    vector_size=50,
    min_count=10,
    workers=2
)


import numpy as np
X = np.array([Cbow.wv.get_mean_vector(mots) for mots in splited_data])
y = df_clean['label'] 

# 6. Split pour l'entraînement
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, random_state=42, stratify=y, test_size=0.2
)

# 7. Lancement XGBoost
from xgboost import XGBClassifier
model_xgb_cbow = XGBClassifier(tree_method='hist')
model_xgb_cbow.fit(X_train, y_train)

print(f"Entraînement terminé ! Accuracy : {model_xgb_cbow.score(X_test, y_test):.4f}")
```
